```python
import os
from os import listdir
import pandas as pd
import json
import time
import numpy as np
```

```python
def get_file(path,vec_path,vec_id):
    files = listdir(path)
    for file_ in files:
        if len(file_.split('.'))==2:
            vec_path.append(path+'/'+file_)
            vec_id.append(file_)
        else:
            get_file(path+'/'+file_,vec_path,vec_id)
    return vec_path    
```

```python
n_tags = 50
path = './lastfm_train/'
tags = pd.read_table('./lastfm_top_tags.txt',sep='\t',nrows=n_tags,names=['tag','occurrence'])
train_path = []
vec_id = []
%time train_path = get_file(path,train_path,vec_id)
```
